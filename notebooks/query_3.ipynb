```python
from lxml import etree
import numpy as np
import pandas as pd
```

```python
def extract_duration(s):
    c=0
    duration=''
    for i in list(s):
        if(i==':'):
            c+=1
        if(c==2):
            duration+=i
        if(c==3):
            break
    return duration[1:]
```

```python
import subprocess
host='ACCESS-DENIED\SQLEXPRESS'

q3_time=[]
for i in range(50):
    durations=[]
    path="G:\dbquery-optimizer\query-variants\query_3\q_three_"+str(i+1)+".sql"
    for j in range(10):
        p=subprocess.Popen(['sqlcmd','-S',host,'-i',path,'-p1'],stdout=subprocess.PIPE)
        stdout=p.communicate()[0]
        # print(stdout)
        f_index=str(stdout).find('affected')
        l_index=f_index+25
        s=str(stdout[f_index+3:l_index])
        d=int(extract_duration(s))
        durations.append(d)
    q3_time.append(durations)

q3_time=np.array(q3_time)
q3_time=pd.DataFrame(q3_time)
q3_time
```

```python
import re

def make_template(file,params,replacedParams):
    tree=etree.parse(file)
    xmlstr=etree.tostring(tree)
    xmlstr=xmlstr.decode("utf-8")
    for i in range(len(params)):
        xmlstr,count=re.subn(params[i],replacedParams[i],xmlstr)
        print("{} occurences replaced".format(count))
    xmlstr=xmlstr.replace("'","''")
    return xmlstr


def use_template(file,params,replacedParams):
    tree=etree.parse(file)
    xmlstr=etree.tostring(tree)
    xmlstr=xmlstr.decode("utf-8")
    for i in range(len(params)):
        xmlstr,count=re.subn(params[i],replacedParams[i],xmlstr)
        # print("{} occurences replaced".format(count))
    return xmlstr

def extract_parameters(query):
    params=[]
    s=query
    s_index=query.find("cn.country_code ='")+len("cn.country_code ='")
    param1=''
    while(s[s_index]!="'"):
        param1+=s[s_index]
        s_index+=1
    params.append(param1)
    s_index=query.find("ct.kind ='")+len("ct.kind ='")
    param2=''
    while(s[s_index]!="'"):
        param2+=s[s_index]
        s_index+=1
    params.append(param2)
    s_index=query.find("it.info ='")+len("it.info ='")
    param3=''
    while(s[s_index]!="'"):
        param3+=s[s_index]
        s_index+=1
    param3="''"+param3+"''"
    params.append(param3)
    s_index=query.find("kt.kind ='")+len("kt.kind ='")
    param4=''
    while(s[s_index]!="'"):
        param4+=s[s_index]
        s_index+=1
    param4="''"+param4+"''"
    params.append(param4)
    return params
```

```python
import subprocess
import os
host='ACCESS-DENIED\SQLEXPRESS'
folder=r"G:\dbquery-optimizer\new_query_3"
directory = os.fsencode(folder)
q3p_time=[]  
for file in os.listdir(directory):
    path=os.path.join(directory,file)
    path=path.decode("utf-8")
    durations=[]
    for j in range(5):
        p=subprocess.Popen(['sqlcmd','-S',host,'-i',path,'-p1'],stdout=subprocess.PIPE,stderr=subprocess.PIPE)
        stdout, stderr=p.communicate()
        try:
            f_index=str(stdout).find('affected')
            l_index=f_index+25
            s=str(stdout[f_index+3:l_index])
            d=int(extract_duration(s))
            durations.append(d)
        except:
            print(stderr)
    q3p_time.append(durations)

q3p_time=np.array(q3p_time)
q3p_time=pd.DataFrame(q3p_time)
q3p_time
```
