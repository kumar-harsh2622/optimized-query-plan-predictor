```python
from lxml import etree
import numpy as np
import pandas as pd
```

```python
def extract_duration(s):
    c=0
    duration=''
    for i in list(s):
        if(i==':'):
            c+=1
        if(c==2):
            duration+=i
        if(c==3):
            break
    return duration[1:]
```

```python
import subprocess
host='ACCESS-DENIED\SQLEXPRESS'

q2_time=[]
for i in range(50):
    durations=[]
    path="G:\dbquery-optimizer\query-variants\query_2\q_two_"+str(i+1)+".sql"
    for j in range(10):
        p=subprocess.Popen(['sqlcmd','-S',host,'-i',path,'-p1'],stdout=subprocess.PIPE)
        stdout=p.communicate()[0]
        # print(stdout)
        f_index=str(stdout).find('affected')
        l_index=f_index+25
        s=str(stdout[f_index+3:l_index])
        d=int(extract_duration(s))
        durations.append(d)
    q2_time.append(durations)

q2_time=np.array(q2_time)
q2_time=pd.DataFrame(q2_time)
q2_time
```

```python
import re

def make_template(file,params,replacedParams):
    tree=etree.parse(file)
    xmlstr=etree.tostring(tree)
    xmlstr=xmlstr.decode("utf-8")
    for i in range(len(params)):
        xmlstr,count=re.subn(params[i],replacedParams[i],xmlstr)
        print("{} occurences replaced".format(count))
    xmlstr=xmlstr.replace("'","''")
    return xmlstr


def use_template(file,params,replacedParams):
    tree=etree.parse(file)
    xmlstr=etree.tostring(tree)
    xmlstr=xmlstr.decode("utf-8")
    for i in range(len(params)):
        xmlstr,count=re.subn(params[i],replacedParams[i],xmlstr)
        # print("{} occurences replaced".format(count))
    return xmlstr

def extract_parameters(query):
    params=[]
    s=query
    s_index=query.find("it.info ='")+len("it.info ='")
    param1=''
    while(s[s_index]!="'"):
        param1+=s[s_index]
        s_index+=1
    params.append(param1)
    s_index=query.find("mi_idx.info > '")+len("mi_idx.info > '")
    param2=''
    while(s[s_index]!="'"):
        param2+=s[s_index]
        s_index+=1
    params.append(param2)
    s_index=query.find('t.production_year > ')+len('t.production_year > ')
    param3=''
    while(s[s_index]!='\n'):
        param3+=s[s_index]
        s_index+=1
    params.append(param3)
    return params
```

```python
import subprocess
import os
host='ACCESS-DENIED\SQLEXPRESS'
folder=r"G:\dbquery-optimizer\new_query_2"
directory = os.fsencode(folder)
q2p_time=[]  
for file in os.listdir(directory):
    path=os.path.join(directory,file)
    path=path.decode("utf-8")
    durations=[]
    for j in range(5):
        p=subprocess.Popen(['sqlcmd','-S',host,'-i',path,'-p1'],stdout=subprocess.PIPE,stderr=subprocess.PIPE)
        stdout, stderr=p.communicate()
        try:
            f_index=str(stdout).find('affected')
            l_index=f_index+25
            s=str(stdout[f_index+3:l_index])
            d=int(extract_duration(s))
            durations.append(d)
        except:
            print(stderr)
    q2p_time.append(durations)

q2p_time=np.array(q2p_time)
q2p_time=pd.DataFrame(q2p_time)
q2p_time
```
